```python
class State:
    def __init__(self, monkey_pos, box_pos, on_box, has_banana):
        self.monkey_pos = monkey_pos
        self.box_pos = box_pos
        self.on_box = on_box
        self.has_banana = has_banana

    def is_goal(self):
        return self.has_banana

    def __str__(self):
        return f"(Monkey: {self.monkey_pos}, Box: {self.box_pos}, OnBox: {self.on_box}, HasBanana: {self.has_banana})"


def monkey_banana():
    # Initial State
    state = State(monkey_pos="door", box_pos="window", on_box=False, has_banana=False)

    print("Initial State:", state)

    # Step 1: Monkey goes to box
    if state.monkey_pos != state.box_pos:
        print("Monkey walks to box")
        state.monkey_pos = state.box_pos

    # Step 2: Monkey pushes box to middle
    if state.box_pos != "middle":
        print("Monkey pushes box to middle")
        state.monkey_pos = "middle"
        state.box_pos = "middle"

    # Step 3: Monkey climbs box
    if not state.on_box:
        print("Monkey climbs onto the box")
        state.on_box = True

    # Step 4: Monkey grasps banana
    if state.on_box and state.monkey_pos == "middle":
        print("Monkey grasps the banana")
        state.has_banana = True

    print("Final State:", state)

    if state.is_goal():
        print("Monkey got the banana!")
    else:
        print("Monkey failed to get the banana.")


# Run the program
monkey_banana()
```
